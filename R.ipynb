```r
#Movie Revenue "Crash" (2005)
```

```r
x<-c(1,2,3,4,5,6,7,8,9)
G<-c(9.11,10.50,8.04,8.55,4.73,3.37,2.12,1.78,1.49)
R<-c(9.11,19.61,27.65,36.20,40.93,44.30,46.42,48.20,49.69)
```

```r
plot(x,R,col="blue", main="'Crash' Total Revenue versus Weeks in Release (2005)",xlab="Weeks in Release", ylab="Total Revenue (in millions")
```

```r
relation_nl<-nls(R~c/(1 + a*exp(-b*x)), start=list(c=90,a=1,b=0.1))
print(relation_nl)
summary(relation_nl)
```

```r
x_seq <- seq(from=1,to=9, by=0.01)
new_relation_nl<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation_nl[i]<-predict(relation_nl, x_new)
}
plot(x,R,col="blue", main="Nonlinear Regression Example",lines(x_seq, new_relation_nl),xlab="t",ylab="R")
```

```r
RSE_Percent = 1.15/mean(R) * 100
```

```r
RSE_Percent
```

```r
confint(relation_nl, level=0.95)
```

```r
plot(fitted(relation_nl), residuals(relation_nl))
```

```r
plot(x,G,col="blue", main="'Crash' Gross Revenue versus Weeks in Release (2005)",xlab="Weeks in Release", ylab="Gross Revenue (in millions")
```

```r
relation_nl2<-nls(G~a*exp(-k*x), start=list(a=8,k=0.1))
print(relation_nl2)
summary(relation_nl2)
```

```r
x_seq <- seq(from=1,to=9, by=0.01)
new_relation_nl2<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation_nl2[i]<-predict(relation_nl2, x_new)
}
plot(x,G,col="blue", main="Nonlinear Regression Example",lines(x_seq, new_relation_nl2),xlab="t",ylab="G")
```

```r
RSE_Percent2
```

```r
plot(fitted(relation_nl2), residuals(relation_nl2))
```
